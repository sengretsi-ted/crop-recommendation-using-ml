```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../input/c-r-dataset/Crop Recommendation Dataset 01.csv')
```

```python
# to give a finite list of values to 'label' columns and putting it into a new column 'target'
c = df.label.astype('category')

# assigning a list and its corresponding index to a variable 'target' using 'enumerate'
# dict for key-value pairs to identify data where the value is the data (using the key) 
targets = dict(enumerate(c.cat.categories))

# assigning 'c' to 'target' with increasing each unique value an incremented integer value
df['target'] = c.cat.codes

# assigning target to 'y'
y = df.target

# assigning features to 'X'
X = df[['N','P','K','temperature','humidity','ph','rainfall']]
```

```python
# importing train and test split
from sklearn.model_selection import train_test_split

# importing MinMax Scaler for scaling the minimum and maximum values to a range
from sklearn.preprocessing import MinMaxScaler

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=1, test_size = 0.2)

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

# we must apply the scaling to the test set as well that we are computing for the training set
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn import metrics
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()

lg.fit(X_train, y_train)

prediction = lg.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, prediction))
```

```python
from sklearn.tree import DecisionTreeClassifier
new = DecisionTreeClassifier(max_depth=10, random_state=1)

new.fit(X_train, y_train)

newm = new.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, newm))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()

nb.fit(X_train, y_train)

pred = nb.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, pred))
```

```python
from sklearn import svm
svm_model = svm.SVC()

svm_model.fit(X_train, y_train)
pred2 = svm_model.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, pred2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors = 5)

knn_model.fit(X_train, y_train)
pred3 = knn_model.predict(X_test)

# print("Accuracy:", metrics.accuracy_score(y_test, pred3))
knn_model.score(X_test, y_test)
```

```python
from sklearn.neural_network import MLPClassifier
mlp_model = MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)

mlp_model.predict_proba(X_test)
lpred = mlp_model.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, pred3))
```
